```python
import os
import matplotlib.pyplot as plt
```

```python
# Set dataset Directory
Normal_dir = "../TB_Chest_Radiography_Database/Normal/"
Tuberculosis_dir = "../TB_Chest_Radiography_Database/Tuberculosis/"
```

```python

# Get image paths from dataset directory
Normal_paths = [os.path.join(Normal_dir, img) for img in os.listdir(Normal_dir)]


# Get image paths from dataset directory
Tuberculosis_paths = [os.path.join(Tuberculosis_dir, img) for img in os.listdir(Tuberculosis_dir)]
```

```python
# Set the style to default
plt.style.use('default')
```

```python
# Count the number of images in each category
Normal_count = len(Normal_paths)
Tuberculosis_count = len(Tuberculosis_paths)

# Create histograms
plt.bar(['Normal', 'Tuberculosis'], [Normal_count, Tuberculosis_count])

# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Number of Images')
plt.title('Distribution of Images in Each Category')

# Display the plot
plt.show()
```
